```python
import pandas as pd
import numpy as np
data={'Product':['Laptop','Moblie','Tablet','Headphones','Smartwatch','Camera'],'Price':[55000,20000,30000,2000,8000,45000],'Units_Sold':[25,60,35,80,50,15,]}
df=pd.DataFrame(data)
print(df)
```

```python
#Data Binning
bin1=[0,10000,30000,60000]
labels1=['Low','Medium','High']
df['Price_Category']=pd.cut(df['Price'],bins=bin1,labels=labels1)
print(df)
```

```python
#Binning Units Sold
bins2=[0,30,60,100]
labels2=['Low Sales','Medium Sales','High Sales']
df['Sales_Category']=pd.cut(df['Units_Sold'],bins=bins2,labels=labels2)
print(df)
```

```python
#Data Formatting
#Convert Data
print(df.dtypes)
df['Units_Sold']=df['Units_Sold'].astype(float)
print(df.dtypes)
print(df)
```

```python
#Convert Text to Uppercase
df['Product']=df['Product'].str.upper()
print(df)
```

```python
#Convert Text to Lowercase
df['Product']=df['Product'].str.lower()
print(df)
```

```python
#Rounding Values
df['Price']=df['Price'].round(2)
print(df)
```

```python
#Additional Operations
#Sort Data
sorted_df=df.sort_values(by='Price',ascending=True)
print(sorted_df)
```

```python
df.sort_values(by='Units_Sold',ascending=False,inplace=True)
print(df)
```

```python
#Display Uniques Values
print(df['Price_Category'].unique())
```

```python
#Dataset: Food Delivery Orders
import pandas as pd
import numpy as np
data={'Order_ID':['01','02','03','04','05','06','07','08'],'Order_Value':[250,450,120,600,300,150,500,220],'Delivery_Time':[30,45,25,60,35,20,50,28],'Distance_km':[5,8,3,10,6,2,9,4]}
df=pd.DataFrame(data)
print(df)
```

```python
bin1=[0,200,400,1000]
labels1=['Low Cost','Medium Cost','High Cost']
df['Order_Category']=pd.cut(df['Order_Value'],bins=bin1,labels=labels1)
print(df)
```

```python
bin2=[0,30,45,70]
labels2=['Fast','Intermediate','Slow']
df['Delivery_Rapidness']=pd.cut(df['Delivery_Time'],bins=bin2,labels=labels2)
print(df)
```

```python
bin3=[0,4,6,11]
labels3=['Close','Moderate','Far']
df['Delivery_Distance']=pd.cut(df['Distance_km'],bins=bin3,labels=labels3)
print(df)
```
